<h5>Qual equipe teve a maior pontuação média por partida ao longo dos anos?</h5>

```python
brasileirao.head()
```

```python
media_classificacao = brasileirao[(brasileirao['ano'] >= 2003) & (brasileirao['ano'] <= 2022)].groupby('equipe')['posicao'].mean()
media_classificacao
```

<h5>Quais equipes tiveram o pior desempenho geral durante o período de 2003 a 2022?</h5>

```python
pior_desempenho = pd.DataFrame(media_classificacao.sort_values(ascending=False))
pior_desempenho
```

<h5>Qual foi o ano com o maior número de empates?</h5>

```python
brasileirao
```

```python
maior_ano_empates = brasileirao.groupby('ano')['empates'].sum().idxmax()
maior_ano_empates
```

```python
total_empates_por_ano = brasileirao.groupby('ano')['empates'].sum().reset_index()
total_empates_por_ano
```

```python
total_empates_por_ano = total_empates_por_ano.rename(columns={'empates': 'total_empates'}).sort_values(by='total_empates', ascending=False)
total_empates_por_ano
```

<h5>Houve alguma tendência específica ao longo dos anos, como aumento ou diminuição do número de gols marcados por partida?</h5>

```python
import matplotlib.pyplot as plt
```

```python
media_gols_por_ano = brasileirao.groupby('ano')['gols_por_partida'].mean().reset_index()

# Crie um gráfico de linha para visualizar as tendências ao longo do tempo
plt.figure(figsize=(10, 6))
plt.plot(media_gols_por_ano['ano'], media_gols_por_ano['gols_por_partida'], marker='o', linestyle='-', color='b')
plt.title('Média de Gols por Partida ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Média de Gols por Partida')
plt.grid(True)
plt.show()
```

<h5>Times mais vitoriosos entre 2009 e 2018</h5>

```python
periodo = (brasileirao['ano'] >= 2009) & (brasileirao['ano'] <= 2018)
dados_periodo = brasileirao[periodo]
```

```python
vitorias_por_equipe = dados_periodo.groupby('equipe')['vitorias'].sum().reset_index()
```

```python
vitorias_por_equipe = vitorias_por_equipe.sort_values(by='vitorias', ascending=False)
vitorias_por_equipe
```

```python
plt.figure(figsize=(12, 6))
plt.bar(vitorias_por_equipe['equipe'], vitorias_por_equipe['vitorias'])
plt.title('Times Mais Vitoriosos entre 2009 e 2018')
plt.xlabel('Equipe')
plt.ylabel('Total de Vitórias')
plt.xticks(rotation=45, ha="right")  # Rotação dos rótulos no eixo x para facilitar a leitura
plt.tight_layout()
plt.show()
```

<h5>Maior média de Gols por Temporada</h5>

```python
media_gols_temporada = brasileirao[{'ano', 'equipe', 'gols_marcados'}]
media_gols_temporada = media_gols_temporada.groupby('ano').mean().sort_values('gols_marcados', ascending= False).rename(columns={'gols_marcados':
    
    'Média de Gols'})
media_gols_temporada['Média de Gols'] = media_gols_temporada['Média de Gols'].round(2)
media_gols_temporada
```

<h5>Times com melhor ataque nos últimos anos</h5>

```python
melhor_ataque = brasileirao[{'ano', 'equipe', 'gols_marcados'}]
melhor_ataque = melhor_ataque.sort_values('gols_marcados', ascending= False)
```

```python
melhor_ataque.head(10)
```

<h5>Adicionar uma coluna de media de gols por jogo no DataFrame acima</h5>

```python
melhor_ataque['média por jogo'] = melhor_ataque['gols_marcados'] / brasileirao['jogos']
```

```python
melhor_ataque = melhor_ataque.round(4)
melhor_ataque.head(10)
```

```python
ordenar_por_media = melhor_ataque.sort_values('média por jogo', ascending=False)
ordenar_por_media.head(10)
```
