<h5>Contagem de vitorias por equipe</h5>

```python
contagem_vitorias = brasileirao.groupby('equipe')['vitorias'].sum()
contagem_vitorias
```

```python
contagem_vitoriasDF = pd.DataFrame(contagem_vitorias)
contagem_vitoriasDF = pd.DataFrame(contagem_vitorias.sort_values(ascending=False))
contagem_vitoriasDF
```

<h5>Qual equipe venceu o campeonato mais vezes ao longo dos anos?</h5>

```python
brasileirao.head()
```

```python
campeoes = brasileirao.loc[brasileirao['posicao'] == 1]
campeoes
```

```python
maiores_campeoes = pd.DataFrame(campeoes['equipe'][campeoes['posicao'] == 1].value_counts())
maiores_campeoes
```

<h5>Qual equipe teve o desempenho mais consistente ao longo dos anos?</h5>

```python
brasileirao
```

```python
equipes_consistentes = pd.DataFrame(brasileirao.groupby('equipe')['posicao'].mean().sort_values())
equipes_consistentes
```
